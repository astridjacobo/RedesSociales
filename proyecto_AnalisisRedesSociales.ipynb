## Procesar uso de redes sociales

### Subtask:
Limpiar y normalizar la columna 'Uso de Redes Sociales' para poder contar las redes sociales individuales. Esto incluirá dividir las cadenas por el delimitador (';'), eliminar espacios extra y manejar posibles entradas duplicadas dentro de una misma celda.


```python
def clean_social_media_entry(entry):
    if pd.isna(entry):
        return []
    networks = entry.split(';')
    cleaned_networks = []
    for net in networks:
        cleaned_net = net.strip().lower()
        # Handle misspellings and variations
        if 'whatszap' in cleaned_net or 'whatsapp' in cleaned_net:
            cleaned_net = 'whatsapp'
        elif 'x antigo twitter' in cleaned_net or 'x (twitter)' in cleaned_net:
            cleaned_net = 'x (twitter)'
        elif 'brawl stars' in cleaned_net:
            cleaned_net = 'brawl stars'
        elif 'telegram e pinterest' in cleaned_net:
            # Split composite entry into individual networks
            cleaned_networks.append('telegram')
            cleaned_networks.append('pinterest')
            continue
        cleaned_networks.append(cleaned_net)
    return cleaned_networks

print("Function 'clean_social_media_entry' defined.")
```

```python
df_redesSociales['cleaned_social_media'] = df_redesSociales['Uso de Redes Sociales'].apply(clean_social_media_entry)

all_social_networks = []
for index, row in df_redesSociales.iterrows():
    all_social_networks.extend(row['cleaned_social_media'])

all_social_networks_series = pd.Series(all_social_networks)

print("Aplicación de la función de limpieza y extracción de redes sociales completada.")
print("Primeros 10 elementos de la serie de todas las redes sociales:")
print(all_social_networks_series.head(10))
```

```python
social_network_counts = all_social_networks_series.value_counts()
most_popular_social_network = social_network_counts.index[0]

print("Frecuencia de cada red social:")
print(social_network_counts)
print(f"\nLa red social más popular es: {most_popular_social_network}")
```

## Extraer mínimo de horas de uso de redes sociales

### Subtask:
Crear una función para parsear la columna 'Tiempo medio diario de uso de redes sociales' y extraer el valor numérico mínimo de horas para cada entrada, manejando diferentes formatos (ej. 'Menos de 1 hora', 'Entre X y Y horas', 'Más de Z horas', y entradas combinadas).


```python
import numpy as np

def extract_min_hours(time_str):
    if pd.isna(time_str) or not isinstance(time_str, str): # Handle NaN or non-string inputs
        return None

    parts = time_str.split(';')
    min_overall = float('inf') # Initialize with a very large number

    for part in parts:
        part = part.strip() # Remove leading/trailing whitespace
        current_min = None

        if 'Menos de 1 hora' in part:
            current_min = 0
        elif 'Entre' in part and 'y' in part and 'horas' in part:
            try:
                # Example: 'Entre 1 y 3 horas'
                num_str = part.split('Entre ')[1].split(' y ')[0].strip()
                current_min = float(num_str)
            except (ValueError, IndexError):
                pass # Keep current_min as None if parsing fails
        elif 'Más de' in part and 'horas' in part:
            try:
                # Example: 'Más de 5 horas'
                num_str = part.split('Más de ')[1].split(' horas')[0].strip()
                current_min = float(num_str)
            except (ValueError, IndexError):
                pass # Keep current_min as None if parsing fails

        # Update overall minimum if a valid current_min is found
        if current_min is not None:
            min_overall = min(min_overall, current_min)

    return min_overall if min_overall != float('inf') else None

print("Function 'extract_min_hours' defined.")
```

```python
df_redesSociales['Minimo de Horas de Uso'] = df_redesSociales['Tiempo medio diario de uso de redes sociales'].apply(extract_min_hours)

print("Primeras filas del DataFrame con la nueva columna 'Minimo de Horas de Uso':")
display(df_redesSociales.head())

print("\nConteo de valores en la columna 'Minimo de Horas de Uso':")
print(df_redesSociales['Minimo de Horas de Uso'].value_counts(dropna=False))
```
